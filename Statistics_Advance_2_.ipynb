**What is hypothesis testing in statistics?**

**Ans.**

Hypothesis testing is a cornerstone of inferential statistics, used to make decisions or inferences about population parameters based on sample data.
The process involves formulating two hypotheses.

1. **Null Hypothesis (H‚ÇÄ):** This is a statement of no effect or no difference. It asserts that any observed effect in the data is due to sampling variability or chance.

2. **Alternative Hypothesis (H‚ÇÅ or Ha):** This posits that there is a genuine effect or difference, contradicting the null hypothesis.

**What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Ans.**

The null hypothesis (denoted as H‚ÇÄ) and the alternative hypothesis (denoted as H‚ÇÅ or Ha) are fundamental concepts in hypothesis testing.

**Null Hypothesis (H‚ÇÄ)**

* **Definition:** This is a statement of no effect, no difference, or no relationship between variables. It assumes that any observed differences in the data are due to chance or random variation.

* **Example:** Suppose we want to test if a new drug is more effective than a standard drug. The null hypothesis would be: "There is no difference in effectiveness between the new drug and the standard drug."

**Alternative Hypothesis (H‚ÇÅ or Ha)**

* **Definition:** This is a statement that contradicts the null hypothesis. It represents the effect, difference, or relationship that we suspect or want to test for.

* **Example:** In the same drug study, the alternative hypothesis would be: "The new drug is more effective than the standard drug."

**Differences**

* **Purpose:** The null hypothesis serves as a default or starting assumption. The alternative hypothesis is what you are trying to provide evidence for.

* **Testing:** The goal of hypothesis testing is to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

* **Interpretation:** If the evidence (usually in the form of a test statistic and p-value) is strong enough, the null hypothesis is rejected, suggesting that the alternative hypothesis is more likely to be true

**3. What is the significance level in hypothesis testing, and why is it important?**

3. **Standardization:** Using common significance levels (e.g., 0.05) allows for consistency and comparability across studies, making it easier to communicate and interpret results.

**4. What does a P-value represent in hypothesis testing?**

A p-value, or probability value, is a fundamental concept in hypothesis testing that helps quantify the evidence against the null hypothesis.

**P-value**

* **Definition:** The p-value is the probability of obtaining test results at least as extreme as the results actually observed, under the assumption that the null hypothesis is true. It essentially measures how well the observed data align with the null hypothesis.

* **Interpretation:** A smaller p-value indicates stronger evidence against the null hypothesis. Conversely, a larger p-value suggests that the observed data are consistent with the null hypothesis.

**Importance**

1. **Decision Making:** The p-value helps determine whether to reject the null hypothesis. If the p-value is less than or equal to the pre-determined significance level (Œ±), the null hypothesis is rejected.

2. **Evidence Against H‚ÇÄ:** A low p-value (typically ‚â§ 0.05) suggests that the observed data are unlikely under the null hypothesis, providing evidence in favor of the alternative hypothesis.

3. **Quantifying Uncertainty:** The p-value quantifies the uncertainty associated with the sample data, helping researchers make informed decisions.

**Relationship between Type 1 and Type 2 Errors**

* **Power of the Test:** The power of a statistical test is defined as 1‚àíùõΩ, representing the probability of correctly rejecting a false null hypothesis. Increasing the sample size or the effect size can increase the power of the test.

There are several types of ANOVA tests, each designed for different experimental designs and types of data.

1. **One-Way ANOVA:**

* Purpose: To compare the means of three or more independent groups based on one independent variable (factor).

* Example: Testing the effect of different teaching methods on student performance.

2. **Two-Way ANOVA:**

* Purpose: To compare the means of groups based on two independent variables (factors) and to study the interaction between these factors.

* Example: Testing the effect of different teaching methods and different times of day on student performance.

3. **Repeated Measures ANOVA:**

* Purpose: To compare means across multiple measurements taken from the same subjects over time or under different conditions.

* Example: Testing the effect of a drug on blood pressure measured at different time points.

4. **Mixed-Design ANOVA:**

* Purpose: Combines features of both between-subjects and within-subjects designs, allowing for the comparison of means across different groups and repeated measures.

* Example: Testing the effect of a drug (between-subjects factor) on blood pressure measured at different time points (within-subjects factor).

5. **MANOVA (Multivariate Analysis of Variance):**

*Purpose: To compare multiple dependent variables across groups based on one or more independent variables.

* Example: Testing the effect of a treatment on multiple health outcomes, such as blood pressure, cholesterol levels, and weight.

6. **ANCOVA (Analysis of Covariance):**

* Purpose: To compare means while controlling for the effects of one or more covariates (continuous variables).

* Example: Testing the effect of a training program on test scores while controlling for initial proficiency levels.

7. **MANCOVA (Multivariate Analysis of Covariance):**

* Purpose: To compare multiple dependent variables across groups while controlling for the effects of covariates.

* Example: Testing the effect of a treatment on multiple health outcomes while controlling for baseline measurements.

8. **Factorial ANOVA:**

* Purpose: To analyze the impact of two or more independent variables (factors) simultaneously and to assess the interaction effects between them.

* Example: Testing the effects of different diets and exercise programs on weight loss.

**22. What is the F-test, and how does it relate to hypothesis testing?**

**Ans.**

The F-test is a statistical test used to compare variances and test hypotheses about them. It is particularly useful in hypothesis testing when comparing two or more samples to understand if they come from populations with equal variances. The F-test is widely applied in various scenarios, including Analysis of Variance (ANOVA) and regression analysis.

**Key Concepts of the F-test:**

1. **Test Statistic:** The F-statistic is the ratio of two sample variances and is calculated as:

$$ F = \frac{ (Variance of Sample 1)} { (Variance of Sample 2)} $$

Here, the larger variance is typically placed in the numerator to ensure the F-value is always ‚â• 1.

2. **F-Distribution:** The test statistic follows an F-distribution, which is right-skewed and determined by the degrees of freedom of the numerator (df‚ÇÅ) and the denominator (df‚ÇÇ).

**Applications of the F-test:**

1. **Analysis of Variance (ANOVA):**

* One-Way ANOVA: Tests whether the means of three or more groups are significantly different by comparing their variances.

* Two-Way ANOVA: Assesses the impact of two independent variables on a dependent variable, including interaction effects.

2. **Regression Analysis:**

* Evaluates the overall significance of a regression model by comparing the explained variance to the unexplained variance.

3. **Comparing Two Variances:**

* Compares the variances of two independent samples to determine if they are significantly different.

* For example, testing if the variances in test scores between two teaching methods are the same.

**Steps in Conducting an F-test:**

1. **State the Hypotheses:**

* Null Hypothesis (H‚ÇÄ): The variances of the two populations are equal (œÉ‚ÇÅ¬≤ = œÉ‚ÇÇ¬≤).

* Alternative Hypothesis (H‚ÇÅ): The variances of the two populations are not equal (œÉ‚ÇÅ¬≤ ‚â† œÉ‚ÇÇ¬≤).

2. **Calculate the F-Statistic:**

* Compute the variances of the two samples.

* Calculate the F-statistic as the ratio of these variances.

3. **Determine the Critical Value:**

* Use the F-distribution table to find the critical value based on the degrees of freedom and the chosen significance level (e.g., 0.05).

4. **Compare and Make a Decision:**

* f the calculated F-statistic is greater than the critical value, reject the null hypothesis, indicating that the variances are significantly different.

* If the F-statistic is less than or equal to the critical value, fail to reject the null hypothesis, indicating that the variances are not significantly different.

**1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**

**Ans.**

**Explanation and Interpretation:**

1. Import Libraries: We import numpy for numerical operations and scipy.stats for statistical functions.

2. Function Definition: We define a function z_test that takes the sample data, population mean, and population standard deviation as inputs and calculates the Z-score and p-value.

3. Sample Data: We use a sample dataset (sample_data) and specify the known population mean and standard deviation.

4. Perform Z-test: We call the z_test function with the sample data, population mean, and population standard deviation.

5. Display Results: We print the Z-score and p-value.

6. Interpretation: We interpret the p-value using a significance level (alpha) of 0.05. If the p-value is less than 0.05, we reject the null hypothesis, indicating that the sample mean is significantly different from the population mean. Otherwise, we fail to reject the null hypothesis.


```python
import numpy as np
import scipy.stats as stats

# Function to perform the Z-test
def z_test(sample_data, population_mean, population_std_dev):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std_dev / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

# Sample data
sample_data = [5.1, 4.9, 5.0, 5.2, 5.3, 5.1, 4.8, 5.0, 5.2, 5.1]
population_mean = 5.0
population_std_dev = 0.2

# Perform Z-test
z_score, p_value = z_test(sample_data, population_mean, population_std_dev)

# Display the results
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpretation of results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

**5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

**Ans.**

**Type 1 and Type 2 Errors:**

**Type 1 Error (Œ±):** The error of rejecting a true null hypothesis (false positive).

**Type 2 Error (Œ≤):** The error of failing to reject a false null hypothesis (false negative).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    # Calculate the critical value and rejection region for the null hypothesis
    z_critical = stats.norm.ppf(1 - alpha / 2)
    standard_error = sigma / np.sqrt(n)
    rejection_region = (mu0 + z_critical * standard_error, mu0 - z_critical * standard_error)

    # Define the range of x values for plotting
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    # Calculate the probability density functions (PDFs)
    pdf_null = stats.norm.pdf(x, mu0, sigma / np.sqrt(n))
    pdf_alternate = stats.norm.pdf(x, mu1, sigma / np.sqrt(n))

    # Plot the PDFs
    plt.plot(x, pdf_null, label="Null Hypothesis (H0)")
    plt.plot(x, pdf_alternate, label="Alternate Hypothesis (H1)")

    # Shade the rejection regions (Type 1 Error)
    plt.fill_between(x, 0, pdf_null, where=(x > rejection_region[0]) | (x < rejection_region[1]), color='red', alpha=0.5, label="Type 1 Error (Œ±)")

    # Shade the overlap region (Type 2 Error)
    plt.fill_between(x, 0, pdf_alternate, where=(x <= rejection_region[0]) & (x >= rejection_region[1]), color='blue', alpha=0.3, label="Type 2 Error (Œ≤)")

    # Labels and title
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.legend()

    # Show the plot
    plt.show()

# Parameters for the test
mu0 = 100       # Null hypothesis mean
mu1 = 105       # Alternate hypothesis mean
sigma = 15      # Population standard deviation
n = 30          # Sample size
alpha = 0.05    # Significance level

# Visualize Type 1 and Type 2 errors
visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha)
```

**9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

**Ans.**

**Explanation of the Function and Its Significance:**

1. Calculate the Sample Mean: The function starts by calculating the sample mean, which is the average of the sample data.

2. Calculate the Standard Error: The standard error is computed by dividing the sample standard deviation by the square root of the sample size.

3. Critical Value: The critical value is determined based on the confidence level. For large sample sizes (typically > 30), the standard normal (Z) distribution is used. For smaller sample sizes, the t-distribution is used.

4. Margin of Error: The margin of error is calculated by multiplying the critical value by the standard error.

5. Confidence Interval: The confidence interval is then computed by adding and subtracting the margin of error from the sample mean.

**Significance of the Confidence Interval:**

A confidence interval provides a range of values within which the true population parameter (in this case, the population mean) is likely to fall, given a certain level of confidence (e.g., 95%). This range gives us an idea of the precision and reliability of our estimate.

For example, if you calculate a 95% confidence interval for a sample mean and it is (48, 52), you can say that you are 95% confident that the true population mean falls within this range. Confidence intervals are crucial in inferential statistics as they provide more information than point estimates and help in making more informed decisions.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    # Calculate the sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the standard error of the mean
    sample_size = len(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)
    standard_error = sample_std_dev / np.sqrt(sample_size)

    # Calculate the critical value (z-score or t-score)
    if sample_size > 30:
        critical_value = stats.norm.ppf((1 + confidence_level) / 2)
    else:
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return ci_lower, ci_upper

# Sample data
sample_data = [45, 50, 55, 49, 52, 48, 53, 47, 51, 49]

# Calculate the confidence interval
ci_lower, ci_upper = confidence_interval(sample_data, confidence_level=0.95)

# Display the results
print(f"Sample Mean: {np.mean(sample_data)}")
print(f"95% Confidence Interval: ({ci_lower}, {ci_upper})")
```

4. Marginal Probability (P(Test Positive)): The total probability of the test being positive, accounting for both true positives and false positives.

5. Posterior Probability (P(Disease | Test Positive)): Using Bayes' Theorem, we update our belief about the probability of the person having the disease given the positive test result.


**12. Perform a Chi-square test for independence between two categorical variables in Python.**

**Ans.**

The Chi-square test for independence is used to determine if there is a significant association between two categorical variables. We'll create a Python program to perform this test using some sample data.



```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Sample data (contingency table)
# Let's consider two categorical variables: Gender (Male, Female) and Preference (Yes, No)
data = pd.DataFrame({
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Female'],
    'Preference': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Count': [20, 30, 25, 25, 30, 20, 35, 15]
})

# Create a contingency table
contingency_table = pd.pivot_table(data, values='Count', index='Gender', columns='Preference', aggfunc=np.sum)
print("Contingency Table:\n", contingency_table)

# Perform Chi-Square Test for Independence
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n {expected}")

# Interpretation of results
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between Gender and Preference.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between Gender and Preference.")
```

**13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.**

**Ans.**

To calculate the expected frequencies for a Chi-square test based on observed data, you'll need to construct a contingency table and then calculate the expected frequency for each cell in the table. The expected frequency for a cell is calculated using the formula:

$$ Expected¬†Frequency = \frac{ (
Row¬†Total√óColumn¬†Total
)} { Within-group¬†variance} $$


```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    # Create a contingency table from observed data
    observed = pd.DataFrame(observed_data)

    # Calculate the row totals and column totals
    row_totals = observed.sum(axis=1).values
    column_totals = observed.sum(axis=0).values
    grand_total = observed.values.sum()

    # Calculate the expected frequencies
    expected = np.outer(row_totals, column_totals) / grand_total

    # Convert to DataFrame for easier readability
    expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)

    return expected_df

# Sample observed data (contingency table)
observed_data = {
    'Category1': [20, 30, 25],
    'Category2': [25, 35, 40]
}
observed_df = pd.DataFrame(observed_data, index=['Group1', 'Group2', 'Group3'])

# Calculate the expected frequencies
expected_df = calculate_expected_frequencies(observed_df)

# Display the observed and expected frequencies
print("Observed Frequencies:\n", observed_df)
print("\nExpected Frequencies:\n", expected_df)
```

A goodness-of-fit test is used to determine if the observed frequencies in a categorical dataset match the expected frequencies. We'll use the Chi-square goodness-of-fit test to compare the observed data to an expected distribution.

```python
import numpy as np
import scipy.stats as stats

# Function to perform the Chi-square goodness-of-fit test
def chi_square_goodness_of_fit(observed, expected):
    # Calculate the Chi-square statistic and p-value
    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
    return chi2_stat, p_value

# Sample observed data
observed = np.array([50, 30, 20])

# Expected frequencies (example: uniform distribution)
total_observed = np.sum(observed)
expected = np.array([total_observed / 3, total_observed / 3, total_observed / 3])

# Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chi_square_goodness_of_fit(observed, expected)

# Display the results
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation of results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies do not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies match the expected distribution.")
```

**Characteristics of the Chi-square Distribution:**

1. Definition: The Chi-square distribution is a continuous probability distribution that arises in hypothesis testing and confidence interval estimation for variance in statistics.

2. Degrees of Freedom (df): The shape of the Chi-square distribution depends on the degrees of freedom. As the degrees of freedom increase, the distribution becomes more symmetrical and approaches a normal distribution.

3. Range: The distribution is only defined for non-negative values (0 to ‚àû) because it represents the sum of the squares of independent standard normal variables.

4. Skewness: The Chi-square distribution is right-skewed, especially for lower degrees of freedom. As the degrees of freedom increase, the skewness decreases.

5. Mean and Variance:

* Mean: The mean of the Chi-square distribution is equal to its degrees of freedom (df).

* Variance: The variance of the Chi-square distribution is 2 * df.

6. Applications: Commonly used in hypothesis testing, such as the Chi-square test for independence in contingency tables and the Chi-square test for goodness of fit. Also used in confidence interval estimation for the population variance of a normal distribution.

**16. Implement an F-test using Python to compare the variances of two random samples.**

**Ans.**

The F-test is used to compare the variances of two populations by examining the ratio of their variances. Let's create a Python script that performs an F-test using two random samples.

```python
import numpy as np
import scipy.stats as stats

# Generate two random samples
np.random.seed(0)  # for reproducibility
sample1 = np.random.normal(loc=10, scale=5, size=50)  # sample from normal distribution with mean 10 and std 5
sample2 = np.random.normal(loc=10, scale=2, size=50)  # sample from normal distribution with mean 10 and std 2

# Calculate the variances of the two samples
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Perform the F-test
f_statistic = var1 / var2
df1 = len(sample1) - 1  # degrees of freedom for sample1
df2 = len(sample2) - 1  # degrees of freedom for sample2
p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

# Print the results
print(f"F-statistic: {f_statistic}")
print(f"Degrees of freedom: {df1} (numerator), {df2} (denominator)")
print(f"P-value: {p_value}")

# Check if the variances are significantly different
alpha = 0.05
if p_value < alpha:
    print("The variances are significantly different.")
else:
    print("The variances are not significantly different.")
```

**19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.**

**Ans.**

1. Normality: Each group should be normally distributed. We can use the Shapiro-Wilk test or visual inspection with Q-Q plots.

2. Independence: Observations should be independent of each other. This is usually ensured by the study design.

3. Equal Variance (Homoscedasticity): The variances of the groups should be equal. We can use Levene's test for this.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):
    # Check Normality using Shapiro-Wilk test
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i + 1} - Shapiro-Wilk Test: Statistics = {stat}, p-value = {p_value}")
        if p_value < 0.05:
            print(f"Group {i + 1} is not normally distributed (p < 0.05)")
        else:
            print(f"Group {i + 1} is normally distributed (p ‚â• 0.05)")

    # Visual inspection with Q-Q plots
    plt.figure(figsize=(10, len(groups) * 5))
    for i, group in enumerate(groups):
        plt.subplot(len(groups), 1, i + 1)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f'Group {i + 1} Q-Q Plot')
    plt.tight_layout()
    plt.show()

    # Check Equal Variance using Levene's test
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test for Equal Variance: Statistics = {stat}, p-value = {p_value}")
    if p_value < 0.05:
        print("The groups do not have equal variance (p < 0.05)")
    else:
        print("The groups have equal variance (p ‚â• 0.05)")

# Example usage
np.random.seed(42)  # for reproducibility
group1 = np.random.normal(loc=5, scale=1, size=30)
group2 = np.random.normal(loc=6, scale=1, size=30)
group3 = np.random.normal(loc=5.5, scale=1, size=30)

check_anova_assumptions(group1, group2, group3)
```

**24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.**



```python
import numpy as np
import scipy.stats as stats

# Simulate a sample from a normal distribution
np.random.seed(42)  # for reproducibility
sample = np.random.normal(loc=10, scale=5, size=50)  # sample with mean 10 and std 5

# Hypothesized population variance
hypothesized_variance = 25  # hypothesized population variance

# Sample variance
sample_variance = np.var(sample, ddof=1)

# Degrees of freedom
df = len(sample) - 1

# Chi-square statistic
chi_square_statistic = (df * sample_variance) / hypothesized_variance

# P-value for the test
p_value = 1 - stats.chi2.cdf(chi_square_statistic, df)

# Print the results
print(f"Sample variance: {sample_variance}")
print(f"Chi-square statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The population variance is significantly different from the hypothesized variance.")
else:
    print("The population variance is not significantly different from the hypothesized variance.")
```

**25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.**



```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(success_a, size_a, success_b, size_b):
    # Proportions
    p1 = success_a / size_a
    p2 = success_b / size_b

    # Combined proportion
    p_combined = (success_a + success_b) / (size_a + size_b)

    # Standard error
    se = np.sqrt(p_combined * (1 - p_combined) * (1 / size_a + 1 / size_b))

    # Z-statistic
    z_statistic = (p1 - p2) / se

    # P-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    return z_statistic, p_value

# Example data
success_a = 30  # Number of successes in group A
size_a = 100  # Total number of trials in group A
success_b = 45  # Number of successes in group B
size_b = 120  # Total number of trials in group B

# Perform Z-test for proportions
z_statistic, p_value = z_test_proportions(success_a, size_a, success_b, size_b)

# Print the results
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference between the proportions of the two groups.")
else:
    print("There is no significant difference between the proportions of the two groups.")
```
